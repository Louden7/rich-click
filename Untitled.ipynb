```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')

# Data: labels and corresponding times
frameworks = [
    'Argparse',
    'Click',
    'rich-click 1.8 üòä',
    'rich-click 1.7 ‚òπÔ∏è',
    'Typer'
]

def make_chart(title, clear_pyc, keep_pyc, ylabel, max_ylim):

    x = np.arange(len(frameworks))  # the label locations
    width = 0.35  # the width of the bars
    
    fig, ax = plt.subplots(figsize=(12,6), dpi=200)
    rects1 = ax.bar(x - width/2, clear_pyc, width, label='Without bytecode', color='#fd7f6f')
    rects2 = ax.bar(x + width/2, keep_pyc, width, label='With bytecode', color='#7eb0d5')
    
    # Adding labels and title
    ax.set_xlabel('\nFramework')
    ax.set_ylabel(ylabel+'\n')
    #ax.set_title('CLI Memory Profiles by Framework')
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(frameworks)
    ax.legend(loc='upper left', facecolor='white')
    ax.set_ylim(0, max_ylim)
    
    # Enhance text label for Rich-Click 1.8
    ax.set_xticks(x)
    ax.get_xticklabels()[2].set_color('black')  # Change Rich-Click 1.8 label color
    ax.get_xticklabels()[2].set_weight('bold')  # Make label bold
    
    # Add some text for labels, title, and custom x-axis tick labels, etc.
    def autolabel(rects):
        """Attach a text label above each bar in *rects*, displaying its height."""
        for rect in rects:
            height = rect.get_height()
            ax.annotate('{:.3f}'.format(round(height, 3)),
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom')
    
    autolabel(rects1)
    autolabel(rects2)
    
    fig.tight_layout()
    return fig
```

```python
clear_pyc_data = [
    0.04577842500000000000,  # Argparse clear .pyc
    0.11450300000000000000,  # Click clear .pyc
    0.24707545000000000000,   # Rich-Click 1.8 clear .pyc
    0.51827485000000000000,   # Rich-Click 1.7 clear .pyc
    0.51097970000000000000,
]
keep_pyc_data = [
    0.04154172500000000000,   # Argparse keep .pyc
    0.06520070000000000000,  # Click keep .pyc
    0.10328477500000000000,  # Rich-Click 1.8 keep .pyc
    0.19543617500000000000,   # Rich-Click 1.8 keep .pyc
    0.18066837500000000000
]

fig = make_chart(
    clear_pyc=clear_pyc_data,
    keep_pyc=keep_pyc_data,
    title='Execution Times by CLI Framework',
    ylabel='Average Execution Time (seconds)',
    max_ylim=0.6
)

    
fig.savefig("execution_times.png")

plt.show()
```

```python
import pandas as pd
import altair as alt

# Data: labels and corresponding times in a DataFrame
data = {
    'Framework': [
        'Argparse', 'Argparse',
        'Click', 'Click',
        'rich-click 1.8 üòä', 'rich-click 1.8 üòä',
        'rich-click 1.7 ‚òπÔ∏è', 'rich-click 1.7 ‚òπÔ∏è',
        'Typer', 'Typer'
    ],
    'Condition': ['Without bytecode']*5 + ['With bytecode']*5,
    'Time': [1.234, 0.567, 1.890, 0.123, 2.345, 0.678, 1.456, 0.789, 1.012, 0.345] # example times
}

df = pd.DataFrame(data)

# Creating the Altair chart
chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('Condition:N', axis=alt.Axis(title='Condition')),
    y=alt.Y('Time:Q', axis=alt.Axis(title='Execution Time\n')),
    color=alt.Color('Condition:N', legend=alt.Legend(title="Bytecode Condition")),
    column=alt.Column('Framework:N', header=alt.Header(title='Framework')),
    tooltip=['Framework', 'Condition', 'Time']
).properties(
    width=200,  # Wider bar groups
    height=300
).configure_scale(
    bandPaddingInner=0.2,
    bandPaddingOuter=0.1  # Adjust the padding to make the chart look better
)

# Customize the appearance
chart = chart.configure_view(strokeWidth=0)
chart = chart.configure_axis(grid=True).configure_view(strokeOpacity=0)
chart = chart.configure_title(fontSize=20)

# You can display the chart using
chart
```
